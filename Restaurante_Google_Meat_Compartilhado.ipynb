```python
#Cardápios. Caso o restaurante queira alterar seu cardápio, basta alterar apenas as listas seguindo a estrutura (item1, valor1, item2, valor2 ...)
cardapio_alacarte = ['Camarão empanado', 40.00, 'Bife à parmeggiana', 30.00, 'Filé de robalo', 34.00, 'Bife à cavalo', 27.00, 'Strogonoff de carne', 26.00, 'vaca atolada', 46.00]
cardapio_bebidas = ['Água', 4.00, 'Água com gás', 4.00, 'Refrigerante', 6.00, 'Suco natural', 7.00, 'Cerveja', 14.00]
cardapio_sobremesa = ['Petit gateau', 25.00, 'Creme de papaia', 20.00,'Torta de limão', 15.00, 'Brigadeiro', 5.00] 



#Informações do carrinho carrinho 
carrinho = []       #carrinho = [quantidade1, nome1, valor1, quantidade2, nome2, valor2 ...]
valor_carrinho = 0  #soma dos valores: valor1 + valor2 + valor3 ...

#Informações da conta
historico_carrinho = []   #conjunto de todos os carrinhos
valor_total = 0           #soma de todos os carrinho 

#Notas
nota50 = 50
nota20 = 20
nota10 = 10
nota5 = 5 
nota2 = 2
nota1 = 1

#Introdução
while True:
  print('RESTAURANTE GOOGLE MEATS \n A --> Acessar \n F --> Finalizar') 
  menu1 = str(input('Digite a letra correspondente à opção desejada: ')).upper()
  print('\n')

  while menu1 != 'F' and menu1 != 'A':                                    
    menu1 = input('Opção inválida. Favor inserir uma opção válida: ').upper()
    print('\n')

# Menu principal

  if menu1 == 'A':
    while True:

      print('''Menu Principal: 
 1 --> Menu à la carte 
 2 --> Menu de bebidas 
 3 --> Menu de sobremesas 
 4 --> Finalizar o pedido 
 5 --> Finalizar a conta
 6 --> Voltar''')  
      
      menu2 = str(input('Digite o número correspondente à opção desejada: '))
      print('\n')
      while menu2!='1' and menu2!='2' and menu2!='3' and menu2!='4' and menu2!='5' and menu2!='6':
        menu2=input('Opção inválida. Favor inserir um número válido: ')
        print('\n')

  # Menu principal - Menu a la carte -------------------------------------------------------------------------------------------------------------------
      if menu2 =='1':
    # Menu principal - Menu a la carte - Escolha do prato a la carte 
        while True:
          itens_cardapio_alacarte = int(len(cardapio_alacarte)/2)
          print('Menu à la carte:')
          for i in range(itens_cardapio_alacarte):
            print(f' {i+1} --> {cardapio_alacarte[i*2]} ---------- R${cardapio_alacarte[(i*2)+1]}0')
          print(f' {itens_cardapio_alacarte+1} --> Voltar')

          a = False

          while True:
            prato = str(input('Digite o número correspondente à opção desejada: '))
            if prato.isnumeric(): 
              prato = int(prato) #converter string em inteiro
              a = True
            else:
              print('Opção inválida.Tente outra vez.')
              print('\n') 
            if a:
              break
        
          if 1 <= prato <= (len(cardapio_alacarte)/2):

      # Menu principal - Menu a la carte - Escolha do prato a la carte - Escolha da quantidade do prato a la carte
            a = False
            while True:
              quantidade = str(input(f'Digite a quantidade desejada de {cardapio_alacarte[(prato-1)*2]} (Digite 0 para desistir): '))
              if quantidade.isnumeric():
                quantidade = int(quantidade)
                a = True
              else:
                print('\n') 
                print('Opção inválida. Tente outra vez.')
                print('\n')
              if a:
                break

            if 0<quantidade<100 :
              print(f'{quantidade} {cardapio_alacarte[(prato-1)*2]} adicionado(s) ao carrinho')

              carrinho.append(quantidade)                                 #adicionando quantidade no carrinho
              carrinho.append(cardapio_alacarte[(prato-1)*2])             #adicionando item ao carrinho
              carrinho.append(cardapio_alacarte[(prato*2)-1])             #adicionando valor no carrinho

              valor_carrinho += cardapio_alacarte[(prato*2)-1]*quantidade #soma valor carrinho: (item * quantidade)
              print('\n')

            elif quantidade == 0 :
              print('\n') 
            elif quantidade>=100:
              print('\n')
              print('Pedido ultrapassou a quantidade máxima permitida.\nA quantidade máxima é de 99 unidades por item.')
              print('\n')
  #Menu principal - Menu a la carte - voltar
          elif prato == (itens_cardapio_alacarte+1):  #opção 6 do menu, sempre um índice a mais do total de itens do cardápio
            print('\n ')
            break

          else:
            print('\n') 
            print('Opção inválida. Tente outra vez.')
            print('\n')

  # Menu principal - Menu de bebidas --------------------------------------------------------------------------------------------------
      elif menu2 == '2':
            # Menu principal - Menu de bebidas - Escolha da bebida 
        while True:
          itens_cardapio_bebidas = int(len(cardapio_bebidas)/2)
          print('Menu de bebidas:') 
          for i in range(itens_cardapio_bebidas):
            print(f' {i+1} --> {cardapio_bebidas[i*2]} ---------- R${cardapio_bebidas[(i*2)+1]}0')
          print(f' {itens_cardapio_bebidas+1} --> Voltar')

          a = False
          while True:
            bebida = str(input('Digite o número correspondente à opção desejada: '))
            if bebida.isnumeric():
              bebida = int(bebida)
              a = True
            else:
              print('Opção inválida. Tente outra vez.')
              print('\n') 
            if a:
              break

        
          if 1 <= bebida <= (len(cardapio_bebidas)/2):

      # Menu principal - Menu de bebidas - Escolha do bebida - Escolha da quantidade do bebida
            a = False
            while True:
              quantidade = str(input(f'Digite a quantidade desejada de {cardapio_bebidas[(bebida-1)*2]} (Digite 0 para desistir): '))
              if quantidade.isnumeric():
                quantidade = int(quantidade)
                a = True
              else:
                print('\n') 
                print('Opção inválida. Tente outra vez.')
                print('\n')
              if a:
                break

            if 0<quantidade<100 :
              print(f'{quantidade} {cardapio_bebidas[(bebida-1)*2]} adicionado(s) ao carrinho')

              carrinho.append(quantidade)                                         #adicionando quantidade no carrinho
              carrinho.append(cardapio_bebidas[(bebida-1)*2])                     #adicionando item ao carrinho
              carrinho.append(cardapio_bebidas[(bebida*2)-1])                     #adicionando valor no carrinho

              valor_carrinho += cardapio_bebidas[(bebida*2)-1]*quantidade         #soma valor carrinho
              print('\n')

            elif quantidade == 0 :
              print('\n')

            elif quantidade>=100:
              print('\n')
              print('Pedido ultrapassou a quantidade máxima permitida.\nA quantidade máxima é de 99 unidades por item.')
              print('\n')

  # Menu principal - Menu de bebidas - voltar
          elif bebida == (itens_cardapio_bebidas+1):  
            print('\n ')
            break
          else:
            print('\n') 
            print('Opção inválida. Tente outra vez.')
            print('\n')

  # Menu principal - Menu de sobremesas -------------------------------------------------------------------------------------------
      elif menu2 == '3':                                                                                                   
    # Menu principal - Menu sobremesas - Escolha da sobremesa 
        while True:
          itens_cardapio_sobremesa = int(len(cardapio_sobremesa)/2)
          print('Menu de sobremesas:') 
          for i in range(itens_cardapio_sobremesa):
            print(f' {i+1} --> {cardapio_sobremesa[i*2]} ---------- R${cardapio_sobremesa[(i*2)+1]}0')
          print(f' {itens_cardapio_sobremesa+1} --> Voltar')

          a = False
          while True:
            sobremesa = str(input('Digite o número correspondente à opção desejada: '))
            if sobremesa.isnumeric():
              sobremesa = int(sobremesa)
              a = True
            else:
              print('Opção inválida. Tente outra vez.')
              print('\n') 
            if a:
              break

        
          if 1 <= sobremesa <= (len(cardapio_sobremesa)/2):

      # Menu principal - Menu sobremesas - Escolha da sobremesa - Escolha da quantidade de sobremesa
            a = False
            while True:
              quantidade = str(input(f'Digite a quantidade desejada de {cardapio_sobremesa[(sobremesa-1)*2]} (Digite 0 para desistir): '))
              if quantidade.isnumeric():
                quantidade = int(quantidade)
                a = True
              else:
                print('\n') 
                print('Opção inválida. Tente outra vez.')
                print('\n')
              if a:
                break
            if 0<quantidade<100 :
              print(f'{quantidade} {cardapio_sobremesa[(sobremesa-1)*2]} adicionado(s) ao carrinho')
              carrinho.append(quantidade)                                      #adicionando quantidade no carrinho
              carrinho.append(cardapio_sobremesa[(sobremesa-1)*2])             #adicionando item ao carrinho
              carrinho.append(cardapio_sobremesa[(sobremesa*2)-1])             #adicionando valor no carrinho

              valor_carrinho += cardapio_sobremesa[(sobremesa*2)-1]*quantidade #soma valor carrinho
              print('\n')

            elif quantidade == 0 :
              print('\n') 
            
            elif quantidade>=100:
              print('\n')
              print('Pedido ultrapassou a quantidade máxima permitida.\nA quantidade máxima é de 99 unidades por item.')
              print('\n')

  #Menu principal - Menu sobremesas - voltar
          elif sobremesa == (itens_cardapio_sobremesa+1):  
            print('\n ')
            break
          else:
            print('\n') 
            print('Opção inválida. Tente outra vez.')
            print('\n')


  # Menu principal - Finalizar pedido --------------------------------------------------------------------------------
      
      elif menu2 == '4':                                                                                          
        while True:    
          print('Seu pedido contém: ')
          itens_carrinho = int(len(carrinho)/3)
          for i in range(itens_carrinho):
            print(f''' {carrinho[i*3]}, {carrinho[(i*3)+1]}, R${carrinho[(i*3)+2]*carrinho[i*3]}0''')
          print(f'Valor total do pedido: R${valor_carrinho}0')
          print('\n')
        
          confirmar = str(input('''Gostaria de confirmar seu pedido?
 C --> Confirmar o pedido 
 L --> Limpar o pedido
 E --> Editar o pedido  
 V --> Voltar ao menu anterior
Digite a letra correspondente à opção desejada: ''')).upper()
          print('\n')

          while confirmar != 'C' and confirmar != 'L' and confirmar != 'E' and confirmar != 'V':                                    
            confirmar = input('Opção inválida. Favor inserir uma opção válida: ').upper()
            
      # Menu principal - Finalizar pedido - Confirmar   
          if confirmar == 'C' and valor_carrinho != 0:
            print('Pedido realizado')

            historico_carrinho += carrinho
            valor_total += valor_carrinho

            valor_carrinho=0
            carrinho = []
            print('\n')
            break

          elif confirmar == 'C' and valor_carrinho == 0:
            print('Opção inválida porque nenhum pedido foi registrado.')
            print('\n')
            break

      # Menu principal - Finalizar pedido - Limpar carrinho
          elif confirmar == 'L' and valor_carrinho != 0:
            acao2 = str(input('''*ATENÇÃO!*
Você está esvaziando este carrinho. Deseja continuar?
 C --> Confirmar 
 R --> Recusar \n
Digite a letra correspondente à opção desejada: ''')).upper()
            print('\n')
    
            while acao2 != 'C' and acao2 != 'R':                                    
              acao2 = input('Opção inválida. Favor inserir uma opção válida: ').upper()
              print('\n')
            
            if acao2 == 'C':
              valor_carrinho=0
              carrinho = []
              print('Carrinho esvaziado com sucesso')
              print('\n')
              break 

            if acao2 == 'R':
              print('\n') 

          elif confirmar == 'L' and valor_carrinho == 0:
            print('Opção inválida porque nenhum pedido foi registrado.')
            print('\n')
            break

      # Menu principal - Finalizar pedido - Editar pedido   
          elif confirmar == 'E' and valor_carrinho != 0:

            print('Seu pedido contém:')
            itens_carrinho = int(len(carrinho)/3)
            for i in range(itens_carrinho):
              print(f' {i+1} --> {carrinho[i*3]}, {carrinho[(i*3)+1]}, R${carrinho[(i*3)+2]*carrinho[i*3]}0')
            print(f'Valor total do pedido: {valor_carrinho}')
            print('\n') 
            
            while True:

              a = False
              while True:
                carrinho_edit = str(input(f'Escolha o número do pedido que deseja alterar \nOBS:Escolha 0 para não fazer alteração: '))
                if carrinho_edit.isnumeric():
                  carrinho_edit = int(carrinho_edit)
                  a = True
                else:
                  print('\n') 
                  print('Opção inválida. Tente outra vez.')
                  print('\n')
                if a:
                  break

              if 0<=carrinho_edit<=(len(carrinho)/3):   #responsavel pela verificação da escolha do item
                break

              else:
                print('Opção inválida. Tente outra vez.')
                print('\n')

            if carrinho_edit > 0:
              acao2 = str(input(f'''Você está excluindo {carrinho[((carrinho_edit*3)-3)]} {carrinho[((carrinho_edit*3)-2)]} no valor de {carrinho[((carrinho_edit*3)-1)]} cada.\n C --> Confirmar \n R --> Recusar \nEscolha uma das opções acima: ''')).upper()
              print('\n')
      
              while acao2 != 'C' and acao2 != 'R':                                    
                acao2 = input('Opção inválida. Favor inserir uma opção válida: ').upper()
                print('\n')

          # Menu principal - Finalizar pedido - Editar pedido - confirmar alterações
              if acao2 == 'C':

                valor_carrinho = (valor_carrinho) - (carrinho[(carrinho_edit*3)-1]) * (carrinho[(carrinho_edit)*3-3])
                del carrinho[((carrinho_edit*3)-3):(carrinho_edit*3)]              

          # Menu principal - Finalizar pedido - Editar pedido - Cancelar alterações    
              elif acao2 == 'r':
                print('\n') 

            elif carrinho_edit == 0:
              print('\n')

          elif confirmar == 'E' and valor_carrinho == 0:
            print('Opção inválida porque nenhum pedido foi registrado.')
            print('\n')
            break

          elif confirmar == 'V':
            break

  # Menu principal - Finalizar conta
      elif menu2 == '5' and valor_total !=0: 
        print('\n')
        print('Confira sua conta:')
        print('\n')
        itens_historico = int(len(historico_carrinho)/3)
        for i in range(itens_historico):
          print(f'''{historico_carrinho[i*3]}, {historico_carrinho[(i*3)+1]}, R${historico_carrinho[(i*3)+2]*historico_carrinho[i*3]}0''')
        print('\n')
        print(f'Valor total da conta: {valor_total}.')
        print('\n')

        acao1 = str(input('Métodos de pagamento: \n 1 --> Pix \n 2 --> Cartão \n 3 --> Dinheiro \n 4 --> Voltar \nEscolha um dos métodos acima: '))
        print('\n')
    
        while acao1 != '1' and acao1 != '2' and acao1 != '3' and acao1 != '4':                                    
          acao1 = input('Opção inválida. Favor inserir uma opção válida: ')
          print('\n') 

    # Menu principal - Finalizar conta - Pix
        if acao1 == '1':   
          print(f'O valor total da conta: R${valor_total}0. \nLeia o QR code abaixo:')
          print('(ilustração)')
          print('\n')
          print('Agradecemos a preferência! Volte sempre.') 
          print('\n')
                    
          break
    # Menu principal - Finalizar conta - Cartão
        elif acao1 == '2':
          print(f'O valor total da conta: R${valor_total}0. \nPara pagamentos em crédito ou débito dirija-se ao caixa.')
          print('\n')
          print('Agradecemos a preferência! Volte sempre.') 
          print('\n')

          break
    # Menu principal - Finalizar conta - Dinheiro
        elif acao1 == '3':
          print(f'Valor total da conta: R${valor_total}0.')

          while True:
            a = False
            while True:
              pagamento = str(input(f'Informe o valor do pagamento em dinheiro: '))
              if pagamento.isnumeric():
                pagamento = int(pagamento)
                a = True
              else:
                print('\n') 
                print('Opção inválida. Tente outra vez.')
                print('\n')
                
              if a:
                break

            if pagamento < valor_total:
              print(f'Valor insuficiente. O valor total da conta foi de R${valor_total}0.')
              print('O valor do pagamento deve ser igual ou superior ao valor da conta.')
            
            else:
              break

          troco = pagamento - valor_total
          trocof = troco
          
          print(f'Seu troco será no valor de {trocof}. Você receberá as seguintes notas:') 

          # troco nota de R$50
          troco_nota50 = troco//nota50
          troco = troco - troco_nota50*nota50
          if troco_nota50 != 0:
            print(f' {troco_nota50} nota(s) de R$50,00.')

          # troco nota de R$20
          troco_nota20 = troco//nota20
          troco = troco - troco_nota20*nota20
          if troco_nota20 != 0:
            print(f' {troco_nota20} nota(s) de R$20,00.')

          # troco nota de R$10
          troco_nota10 = troco//nota10
          troco = troco - troco_nota10*nota10
          if troco_nota10 != 0:
            print(f' {troco_nota10} nota(s) de R$10,00.')

          # troco nota de R$5
          troco_nota5 = troco//nota5
          troco = troco - troco_nota5*nota5
          if troco_nota5 != 0:
            print(f' {troco_nota5} nota(s) de R$5,00.')

          # troco nota de R$2
          troco_nota2 = troco//nota2
          troco = troco - troco_nota2*nota2
          if troco_nota2 != 0:
            print(f' {troco_nota2} nota(s) de R$2,00.')

          # troco nota de R$1    
          troco_nota1 = troco//nota1
          troco = troco - troco_nota1*nota1
          if troco_nota1 != 0:
            print(f' {troco_nota1} moeda(s) de R$1,00.')

          print('\n')

          print('Agradecemos a preferência! Volte sempre.') 
          print('\n')

          break

  # Menu principal - Finalizar conta - Voltar
        elif acao1 == '4':
          print()

      elif menu2 == '5' and valor_total ==0:
        print('Nenhum pedido foi realizado até o momento')
        print('\n')
      elif menu2 == '6' and valor_total ==0:
        break 
      
      elif menu2 == '6' and valor_total !=0:
        print(f'Existe um valor de R${valor_total}0 em aberto.\nPara realizar o pagamento entre no menu "Finalizar a conta".')
        print('\n')

  elif menu1 == 'F':                                                                                                  
    
    if valor_total != 0:
        print('Nota fiscal')
        itens_historico = int(len(historico_carrinho)/3)
        for i in range(itens_historico):
          print(f'''{historico_carrinho[i*3]}, {historico_carrinho[(i*3)+1]}, R${historico_carrinho[(i*3)+2]*historico_carrinho[i*3]}0''')
          
        print(f'Valor total: {valor_total}')
        print('\n')

        print('Agradecemos a preferência! Volte sempre.')

        print('\n')

        historico_carrinho = []
        valor_total = 0

    elif valor_total == 0:
    
      print('Programa finalizado')
      
    break
```
